## <p align="center" style="color:#5D6D7E;">MODELO MACHINE LEARNING</p>

##### DIVISI√ìN DE DATOS

```python
air
```

```python
air["NO2(GT)"]= pd.to_numeric(air["NO2(GT)"],errors="coerce")
```

```python
def niveles_NO2(concentracion_NO2):
    if concentracion_NO2<=50:
        return "Baja"
    elif concentracion_NO2<=100:
        return "Moderada"
    elif concentracion_NO2<=200:
        return "Alta"
    else:
        return "Muy alta"
```

```python
air["NO2_Categoria"]= air["NO2(GT)"].apply(niveles_NO2)
```

#### train y test

```python
x= air.drop(["NO2(GT)", "NO2_Categoria"],axis=1)
y= air["NO2_Categoria"]
x_train, x_test,y_train, y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
threshold= 100  
y_class= np.where(air["NO2(GT)"] >threshold,1,0)
```

```python
# Dividir datos
x_train, x_test, y_train, y_test = train_test_split(x,y_class, test_size=0.2,random_state=42) 
```

```python
# escalado
scaler = StandardScaler()
x_train_sca = scaler.fit_transform(x_train)
x_test_sca= scaler.transform(x_test)
```

```python
# Iniciar y entrenar el modelo RandomForest
model = RandomForestClassifier()
model.fit(x_train_sca, y_train)
```

```python
y_pred= model.predict(x_test_sca)
```

```python
accuracy= accuracy_score(y_test,y_pred)
print("Accuracy en el conjunto de prueba:",accuracy)

print("Matriz de confusi√≥n:")
print(confusion_matrix(y_test,y_pred))

print("Reporte de clasificaci√≥n:")
print(classification_report(y_test,y_pred))
```

```python
models= {
    "Random Forest": RandomForestClassifier(),
    "SVC": SVC(),
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "GradientBoostingClassifier": GradientBoostingClassifier(), 
    "AdaboostClassifier": AdaBoostClassifier()
}

res=[]

for name, model in models.items():
    model.fit(x_train_sca, y_train)
    train_score= model.score(x_train_sca, y_train)
    test_score= model.score(x_test_sca, y_test)
    res.append({"Modelo": name, "Entrenamiento": train_score,"Prueba": test_score})

res_air = pd.DataFrame(res)
```

```python
res_air
```

```python
param_grid={
    "Random Forest": {
        "model": RandomForestClassifier(),
        "param_grid": {
            "n_estimators": [100, 200,300],
            "max_depth": [None, 10, 20,30],
            "min_samples_split": [2, 5,10],
            "min_samples_leaf": [1,2, 4],
            "max_features": ['auto', 'sqrt', 'log2']}},
    "Logistic Regression": {
        "model": LogisticRegression(),
        "param_grid": {
            "C": [0.001, 0.01, 0.1, 1, 10, 100],
            "penalty": ['l1', 'l2']}},
    "SVC": {
        "model": SVC(),
        "param_grid":{
            "C": [0.1, 1, 10],
            "kernel": ['linear','rbf']}}}

for model_name, params in param_grid.items():
    model= params["model"]
    param_grid= params["param_grid"]
    
    grid_search= GridSearchCV(model, param_grid, cv=5,scoring="accuracy",n_jobs=-1)
    grid_search.fit(x_train_sca, y_train)
    
    print(f"Mejores hiperpar√°metros para {model_name}:{grid_search.best_params_}")
```

#### definir, entrenar y evaluar modelos

```python
logisticR= LogisticRegression(C=0.1,penalty="l2")
logisticR.fit(x_train_sca,y_train)
lr_score=logisticR.score(x_test_sca, y_test)
print("Logistic Regression Accuracy:",lr_score)
```

```python
svc =SVC(C=10, kernel='rbf')
svc.fit(x_train_sca,y_train)
svc_score= svc.score(x_test_sca,y_test)
print("SVC Accuracy:",svc_score)
```

```python
random_forest =RandomForestClassifier(max_depth=None, max_features="sqrt",min_samples_leaf=1,min_samples_split=5,n_estimators=300)
random_forest.fit(x_train_sca, y_train)
rf_score= random_forest.score(x_test_sca,y_test)
print("Random Forest Accuracy:",rf_score)
```
